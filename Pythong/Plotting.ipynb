```python
import numpy as np
import matplotlib.pyplot as plt

a = [pow(10, i) for i in range(10)]
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)

line, = ax.plot(a, color='blue', lw=2)

ax.set_yscale('log')
ax.set_xscale('linear')
plt.show()
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
# headers = ['Wavelength', 'Eumelanin', 'Pheomelanin']
df = pd.read_csv("/Users/joeljohnson/Documents/Github/CG_Research/PhysicalModels/AbsorptionValues.csv");
# print(df.head())

# Read in the data
# df = pd.read_csv('AbsorptionValues.csv')
# WaveLength = df.Wavelength
WaveLength = df.iloc[:41,0].to_numpy()
Eumelanin = df.loc[:40,'Eumelanin'].to_numpy()
Pheomelanin = df.loc[:40,'Pheomelanin'].to_numpy()
Baseline = df.loc[:40,'Baseline'].to_numpy()
# print(WaveLength)
# print(Eumelanin)
# print(Pheomelanin)
# print(Baseline)
# Plot the data
plt.plot(WaveLength, Eumelanin)
plt.plot(WaveLength, Pheomelanin)
plt.plot(WaveLength, Baseline)
plt.yscale('log')
plt.xscale('linear')
plt.xlim(400, 700)

plt.ylim(0.015, 1000)



plt.title('Absorption Values')
plt.xlabel('Wavelength')
plt.ylabel('Absorption')
plt.legend(['Eumelanin', 'Pheomelanin', 'Baseline'])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'red')
plt.show()
```
