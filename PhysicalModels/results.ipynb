```python
import math
wl = 400
Cm = 0.01
Ch = 0.01
Bm = 0.01
gamma = 0.9
alpha_oxy = 100
alpha_deoxy = 100
alpha_s = 14.74*math.pow(wl,-0.22)+2.22*math.pow(10,11)*math.pow(wl,-4.0)
alphas = 1.1*math.pow(10,12)*math.pow(wl,-4.0)+73.7*math.pow(wl,-0.22)
print("Scattering")
print(alpha_s)
print(alphas)
print("Absorption Baseline")
alpha_base = 0.0244 + 8.54 * pow(10,-(wl-154)/66.2)*10
alpha_base2 = 0.0244 + 8.53*np.exp(-(wl-154)/66.2)
print(alpha_base)
print(alpha_base2)
print("Absorption Pheomelanin")
alpha_pm = 2.9 * pow(10,14) * pow(wl,-4.75)
print(alpha_pm)
print("Absorbance Eumelanin")

alpha_em = 6.6 * pow(10,10) * pow(wl,-3.33)
print(alpha_em)
print("Absorbance Melanin")
alpha_m = Cm * alpha_em + Ch * alpha_pm
print(alpha_m)
print("Absorbance Epidermis")
alpha_total = Cm*(Bm*alpha_em + (1-Bm)*alpha_pm) + (1-Cm)*alpha_base
print(alpha_total)
print("Absorbance Dermis")
alpha_derm_total = Ch*(gamma*alpha_oxy + (1-gamma)*alpha_deoxy) + (1-Ch)*alpha_base
print(alpha_derm_total)
```
